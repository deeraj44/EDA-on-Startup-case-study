```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("startup_funding.csv")
df.head()
```

```python
df.rename(columns={"Date dd/mm/yyyy":"Date","City  Location":"City","InvestmentnType":"InvestmentType"},inplace=True)
```

```python
df
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df.drop(columns="Remarks",inplace=True)
df.drop(columns="Sr No",inplace=True)
```

```python
df
```

```python
df = df.drop(df[df['Amount in USD'].isnull()].index )
df =df.drop(df[(df['Amount in USD'] == 'Undisclosed') | (df['Amount in USD'] == 'undisclosed')].index )
df['Amount in USD'] = df['Amount in USD'].apply(lambda x: x.replace(',',''))
df['Amount in USD'] = df['Amount in USD'].apply(lambda x: x.replace('+',''))
df= df[df['Amount in USD'].str.replace('.','',1).str.isdigit()]
```

```python
df["Amount in USD"]= df["Amount in USD"].astype(str).astype(float)
```

```python
df["Amount in USD"]
```

```python
plt.figure(figsize=(20,10))
df["Amount in USD"].plot(color="g")
plt.title("Amount Distribution",fontsize=15,color="g")
plt.xlabel("Amount in USD",fontsize=15)
plt.show()
```

<h3 style=color:green>Distribution of Amount in USD</h3>

```python
df['Date'][df['Date']=='12/05.2015'] = '12/05/2015'
df['Date'][df['Date']=='13/04.2015'] = '13/04/2015'
df['Date'][df['Date']=='15/01.2015'] = '15/01/2015'
df['Date'][df['Date']=='22/01//2015'] = '22/01/2015'
df['Date'][df['Date']=='05/072018'] = '05/07/2018'
df['Date'][df['Date']=='01/07/015'] = '01/07/2015'
df['Date'][df['Date']=='\\\\xc2\\\\xa010/7/2015'] = '10/07/2015'
```

```python
df["year_month"] = (pd.to_datetime(df['Date'],
         format='%d/%m/%Y').dt.year*100)+(pd.to_datetime(df['Date'],format='%d/%m/%Y').dt.month)


temp = df['year_month'].value_counts().sort_values(ascending = False).head(10)
print("Number of funding per month in decreasing order(Top 10)\n",temp)
ym = df['year_month'].value_counts()
```

```python
plt.figure(figsize=(20,8))
sns.barplot(ym.index,ym.values,color="r")
plt.xticks(rotation=90)
plt.title("Year month Distribution",fontsize=15,color="r")
plt.xlabel("Year and month of transaction",fontsize=15)
plt.ylabel("No. of transactions made",fontsize=15)
plt.yticks([5,10,15,20,25,30,35,40,45,50,55,60,65])
plt.grid(axis="y")
plt.show()
```

<h3 style=color:red>Distrubution of transactions made over the months</h3>
